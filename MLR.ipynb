```python
for i in range(len(file_list)):
    if file_list[i]=='data\\ToyotaCorolla.csv':
        file=file_list[i]
        
file
```

```python
car_df = pd.read_csv(file, encoding= 'unicode_escape')
#. 의미는 pd가 가지고 있는 read_csv 기능
car_df
```

```python
car_df = car_df.iloc[0:1000] 
```

```python
car_df.info()
```

```python
car_df.columns
```

```python

predictors = ['Age_08_04', 'KM', 'Fuel_Type', 'HP', 'Met_Color', 'Automatic', 'CC', 'Doors', 'Quarterly_Tax', 'Weight']
#X변수에 저장해줄 것들
outcome = 'Price'
#y 에 저장해줄 시리즈
```

```python
X=pd.get_dummies(car_df[predictors],drop_first=True)
```

```python
y=car_df[outcome]
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train_all, X_test, y_train_all, y_test = train_test_split(X, y,test_size=0.2,random_state=1)#random_state는 seed number 같은 sampling유지
```

```python
X_train,X_valid,y_train,y_valid=train_test_split(X_train_all, y_train_all, test_size=0.2, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
car_lm = LinearRegression()
car_lm.fit(X_train, y_train)
```

```python
#print coefficient
print(pd.DataFrame({'Features':X.columns, 'coefficient':car_lm.coef_}))
```

```python
from dmba import regressionSummary
```

```python
regressionSummary(y_train, car_lm.predict(X_train))
```

```python
regressionSummary(y_valid, car_lm.predict(X_valid))
```

```python
import statsmodels.api as sm  
train_X2 = sm.add_constant(X_train)
est = sm.OLS(y_train, X_train)
print(est.fit().summary())
```

```python
# Use predict() to make predictions on a new set
car_lm_pred = car_lm.predict(X_valid)
result = pd.DataFrame({'Predicted': car_lm_pred, 'Actual':y_valid, 'Residual': y_valid - car_lm_pred})
print(result.head(20))

# print performance measures (validation data)
regressionSummary(y_valid, car_lm_pred)
```
