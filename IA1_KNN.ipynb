```python
import pandas as pd

## Reading data  
universal_df = pd.read_csv("UniversalBank.csv")
# delete Zip.Code column. We are not going to use it for our analysis 
universal_df = universal_df.drop(['ID', 'ZIP Code'], axis=1)
# preview universal.df
universal_df.head()
```

```python
# set CreditCard to categorical variable
universal_df = universal_df.astype({'CreditCard':'category'})
universal_df.dtypes
```

```python
# show basic statistics of variables in the dataframe
universal_df.describe()
```

```python
from sklearn import preprocessing

universal_X = universal_df.iloc[:,0:11]
universal_y = universal_df['CreditCard']

universal_X
```

```python
## Scaling
scaler=preprocessing.StandardScaler()
scaled = scaler.fit_transform(universal_X)
universalNorm_X = pd.DataFrame(scaled, columns = universal_X.columns)
universalNorm_X.head()
```

```python
from sklearn.model_selection import train_test_split

## Partitioning 
#---------------
# Q1. Set seed number so that you can get the same traing and test set whenever you create them. 
# Create trainX (60%) and testX (40%) with universalNormX
# Your code here  
#---------------

train_X, test_X, train_y, test_y = train_test_split(universalNorm_X, universal_y,test_size=0.4,random_state=1)


```

```python
## Running KNN with k=5
# use NearestNeighbors and KNeighborsClassifier from scikit-learn to compute knn
from sklearn.neighbors import NearestNeighbors, KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5).fit(train_X, train_y)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
confusion_matrix(test_y, knn.predict(test_X))
confusion_matrix=confusion_matrix(test_y, knn.predict(test_X))
accuracy_score(test_y, knn.predict(test_X))

#---------------
# Q2. What is the error rate, false positive rate, 
# and false negative rate from the confusion matrix?
# 
# Your answer here
TP=confusion_matrix[0][0]
FN=confusion_matrix[0][1]
FP=confusion_matrix[1][0]
TN=confusion_matrix[1][1]

Err=(FN+FP)/(TP+FN+FP+TN)
print(Err)

FPR=FP/(TN+FP)
print(FPR)

```

```python
## Finding best k

results = []
# compute knn for different k on test set.
for k in range(1, 15):
    knn = KNeighborsClassifier(n_neighbors=k).fit(train_X, train_y)
    results.append({'k': k, 'accuracy': accuracy_score(test_y, knn.predict(test_X))})

# Convert results to a pandas data frame
results = pd.DataFrame(results)
print(results)

#---------------
# Q3. What is the best k?
# 
# Your answer here
# k=4 일때가 best K 라고 생각한다. k=5 일 때 accuracy 가 소폭 감소했고, k=6 이상일 때 k=4 보다 accuracy 가
#크긴 하지만 k 값이 커질 수록 
#---------------
```
